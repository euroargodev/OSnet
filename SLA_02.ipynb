```python
#Read SLA files and subsample the Gulf Stream region - July 2021
import datetime as dt
import glob
import netCDF4 as nc
import numpy as np
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
import cmocean
from gsw import sigma0
import numpy.ma as ma

import yaml

import cartopy
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
cartopy.config["data_dir"] = './cartopy_shapefiles'

os.getcwd()
```

```python
%%time
with open(r'/home1/datahome/mlejeune/configuration.yaml') as file:
    configuration = yaml.load(file, Loader=yaml.FullLoader)

    print("File version: ", configuration["version"])
```

```python
# Clean OutPut
delete = False;

if(delete):
    if len(os.listdir(configuration['SLA_path'])) != 0:

         arr = os.listdir(configuration['SLA_path'])
         for x in arr:
             os.remove((configuration['SLA_path'] + x))
             print("delete file: ", (configuration['SLA_path'] + x))

    else:    
        print("Directory is empty")
```

```python
path = configuration['SLA_DATA_path']

for yy in range(configuration['YEARS_1'],configuration['YEARS_2']):
    ds = xr.open_mfdataset(path + str(yy) + '/**/*.nc', combine='nested', concat_dim='time')
    
    ds = ds.assign_coords(lon180=(((ds.longitude + 180) % 360) - 180))  
    ds['longitude'] = ds.lon180
    
    ds = ds.sortby('longitude')
    
    dsub = ds.where((ds['longitude']>=configuration['longitude1_oceans']) & (ds['longitude']<=configuration['longitude2_oceans']) & (ds['latitude']>=configuration['latitude2_oceans']) & 
              (ds['latitude']<=configuration['latitude1_oceans']), drop=True)

    dsub.to_netcdf(configuration["SLA_path"] + '/' + configuration["SLA_prefix"] + str(yy) + '.nc')
    print(yy)
```

```python
#Load back and test
path = configuration['SLA_DATA_path']

yy = configuration['YEARS_1']
ds = xr.open_mfdataset(path + str(yy) + '/**/*.nc', combine='nested', concat_dim='time')

ds = ds.assign_coords(lon180=(((ds.longitude + 180) % 360) - 180))  
ds['longitude'] = ds.lon180
ds = ds.sortby('longitude')

dsub = ds.where((ds['longitude']>=configuration['longitude1_oceans']) & (ds['longitude']<=configuration['longitude2_oceans']) & (ds['latitude']>=configuration['latitude2_oceans']) & 
              (ds['latitude']<=configuration['latitude1_oceans']), drop=True)
print(dsub)
```

```python
#Verif Maps
proj=ccrs.NorthPolarStereo()

land_feature = cfeature.NaturalEarthFeature(
    category='physical', name='land', scale='50m', facecolor=[0.9375, 0.9375, 0.859375])
extent = np.array([17, -30, 50, 85])
subplot_kw = {'projection': proj, 'extent': extent}

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(25, 15), dpi=120, facecolor='w', edgecolor='k', subplot_kw=subplot_kw)
dsub['sla'].isel(time=0).plot(ax=ax[0],
                  cmap=cmocean.cm.balance,center=False,
                  transform=ccrs.PlateCarree(),
                  cbar_kwargs={"orientation": "horizontal", "shrink": 0.7},
                  robust=True)
ax[0].contour(dsub['lon180'],dsub['latitude'],dsub['sla'].isel(time=0),colors = "black",linewidths = 1)
ax[0].add_feature(land_feature, edgecolor='black')

dsub['err'].isel(time=0).plot(ax=ax[1],
                  cmap=cmocean.cm.amp,center=False,
                  transform=ccrs.PlateCarree(),
                  cbar_kwargs={"orientation": "horizontal", "shrink": 0.7},
                  robust=True)
ax[1].contour(dsub['lon180'],dsub['latitude'],dsub['err'].isel(time=0),colors = "black",linewidths = 1)
ax[1].add_feature(land_feature, edgecolor='black')
plt.show()
```
