```python
#Read SLA files and subsample the Gulf Stream region - July 2021
import datetime as dt
import glob
import netCDF4 as nc
import numpy as np
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
import cmocean
from gsw import sigma0
import numpy.ma as ma

import time
import yaml

import cartopy
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER




cartopy.config["data_dir"] = './cartopy_shapefiles'

os.getcwd()
```

```python
with open(r'configuration.yaml') as file:
    configuration = yaml.load(file, Loader=yaml.FullLoader)

    print("File version: ", configuration["version"])
```

```python
# Clean OutPut
confirm = False;
if(confirm) :
    if os.path.exists(configuration['BATHYMETRIE_output_path']):
      os.remove(configuration['BATHYMETRIE_output_path'] + configuration['BATHYMETRIE_FILE_NAME'])

      print("delete file", configuration['BATHYMETRIE_output_path'] + configuration['BATHYMETRIE_FILE_NAME'])
    else:
      print("Nothing to remove") 
```

```python
#Bathymetrie
    
ds = xr.open_dataset(configuration['BATHYMETRIE_path'], engine='netcdf4')
ds = ds.rename({'x': "LONGITUDE", 'y': 'LATITUDE', 'z': 'bathymetry'})

ds = ds.where((ds['LONGITUDE']>=configuration['longitude1_oceans']) & (ds['LONGITUDE']<=configuration['longitude2_oceans']) & (ds['LATITUDE']>=configuration['latitude2_oceans']) & 
              (ds['LATITUDE']<=configuration['latitude1_oceans']), drop=True)

ds.to_netcdf(configuration['BATHYMETRIE_output_path'] + configuration['BATHYMETRIE_FILE_NAME'])
file_name = configuration['BATHYMETRIE_output_path'] + configuration['BATHYMETRIE_FILE_NAME']

time.sleep(5) #security waiting for generate file
print("file of bathymetrie geneate with succes : ", file_name)

#outPut

dsub = xr.open_dataset(configuration['BATHYMETRIE_output_path'] + configuration['BATHYMETRIE_FILE_NAME'])
dsub
```
